````python
import os
from wit import Wit as WitAPI

WIT_ACCESS_TOKEN = os.getenv("WIT_ACCESS_TOKEN")
wit_client = WitAPI(WIT_ACCESS_TOKEN)

audio_fpath = "/output/audio/google/hello.wav"

with open(audio_fpath, 'rb') as audio:
    transcription = ""
    try:
        wit_transcription = wit_client.speech(
            audio, {'Content-Type': 'audio/wav'})

        if wit_transcription != None:
            print(wit_transcription)
            if "text" in wit_transcription:
                transcription = str(wit_transcription["text"])
    except Exception as e:
        print("Could not request results from Wit.ai service; {0}".format(e))
        transcription = ""
````

### 3.5 Wav2Vec2

````python
import soundfile as sf
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
import torch
import gc
from normalise import normalise, tokenize_basic

def normalize_text(text):
    return " ".join(normalise(text, tokenizer=tokenize_basic, verbose=False))

# m_13746_3
# m_10157_2
# m_15494_1
audio_fpath = "/home/user/asr-fuzzing-testing/CrossASRplus/examples/output/augmenter_full/data/audio/google/m_9064_2.wav"

audio_input, _ = sf.read(audio_fpath)
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

# transcribe
input_values = tokenizer(
    audio_input, return_tensors="pt").input_values
# input_values = input_values.to(self.device)

logits = model(input_values).logits
predicted_ids = torch.argmax(logits, dim=-1)
transcription = tokenizer.batch_decode(predicted_ids)[0]

del audio_input, input_values, logits, predicted_ids
torch.cuda.empty_cache()
gc.collect()

print(transcription)
print(normalize_text(transcription))
````
